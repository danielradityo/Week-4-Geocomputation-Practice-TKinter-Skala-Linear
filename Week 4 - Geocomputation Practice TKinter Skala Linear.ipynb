## Skala Linear

```python
# Import Library Tkinter
from tkinter import *

# Membuat Variabel
## a = 0, A = 0
minDatX = 10
minDatY = 5
maxDatX = 80
maxDatY = 70
interval = 20
posXa = 50
posYa = 50
lebar = 200
panjang = 400

# Proses
minskalaX = (minDatX / interval) * interval
maxskalaX = ((maxDatX / interval) + 1) * interval
jlhgrsX = (maxskalaX - minskalaX) / interval + 1
skalaX = (lebar / (maxskalaX - minskalaX))

minskalaY = (minDatY / interval) * interval
maxskalaY = ((maxDatY / interval) + 1) *  interval
jlhgrsY = (maxskalaY - minskalaY) / interval + 1
skalaY = (panjang / (maxskalaY - minskalaY))

# Pembuatan Skala Linear
## Main Window
root = Tk()
root.title('Uji Coba Tkinter')
root.geometry('700x700')

## Canvas
canvas = Canvas(root, height = 700, width = 700, bg = 'white')

### Penentuan Titik X (menggunakan while Loop)
i = 0
while i < jlhgrsX:
    i += 1
    dataX = (interval * i ) + minskalaX
    posX = posXa + (dataX - minskalaX) * skalaX
    line = canvas.create_line(posX, posYa - 5, posX, posYa + 5)
    
### Penentuan Titik Y (menggunakan while Loop)
i = 0
while i < jlhgrsY:
    i += 1
    dataY = (interval * i) + minskalaY
    posY = posYa + (dataY - minskalaY) * skalaY
    line = canvas.create_line(posXa - 5, posY, posXa + 5, posY)
    
### Create Rectangle (sebagai base kotak skala linear)
rectangle = canvas.create_rectangle(posXa, posYa, posX, posY, fill='white')
    
canvas.pack()
root.mainloop()
```

## Codingan C++ (Jangan di run)

```python
#include <iostream> //Doni Maulana Syahputra 118150100 Geokom Shift B
#include <graphics.h> //Skala linear minggu 4
#include <conio.h>

int main () {
initwindow (600, 700, "Skala Linear Doni Maulana S");
rectangle (100, 100, 300, 500);

//deklarasi
// a = 0, A = 0
int minDatX = 10;
int minDatY = 5;
int maxDatX = 80;
int maxDatY = 70;
int interval = 20;
int posXa = 100;
int posYa = 100;
int lebar = 200;
int panjang = 400;

int minskalaX, maxskalaX,jlhgrsX, skalaX, posX, dataX;
int minskalaY, maxskalaY,jlhgrsY, skalaY, posY, dataY;

//proses
minskalaX = (minDatX / interval) * interval;
maxskalaX = ((maxDatX / interval) + 1) * interval;
jlhgrsX = (maxskalaX - minskalaX) / interval + 1;
skalaX = (lebar / (maxskalaX - minskalaX));
posX = posXa + (dataX - minskalaX) * skalaX;

minskalaY = (minDatY / interval) * interval;
maxskalaY = ((maxDatY / interval) + 1) *  interval;
jlhgrsY = (maxskalaY - minskalaY) / interval + 1;
skalaY = (panjang / (maxskalaY - minskalaY));
posY = posYa + (dataY - minskalaY) * skalaY;

//penentuan titik
for(int i = 0; i < jlhgrsX; i++) {
    dataX = (interval * i ) + minskalaX;
    posX = posXa + (dataX - minskalaX) * skalaX;
    line(posX, posYa - 5, posX, posYa + 5);
}
for(int i = 0; i < jlhgrsY; i++) {
    dataY = (interval * i) + minskalaY;
    posY = posYa + (dataY - minskalaY) * skalaY;
    line (posXa - 5, posY, posXa + 5, posY);
}

getch();
closegraph();
return 0;
}
```
